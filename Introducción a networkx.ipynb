```python
er = nx.erdos_renyi_graph(200, 0.02)

ws = nx.watts_strogatz_graph(200, 4, 0.03)

ba = nx.barabasi_albert_graph(200, 4)
```

```python
subplot(1, 3, 1)
nx.draw(er)
axis('image')
title('Erdos-Renyi')

subplot(1, 3, 2)
nx.draw(ws)
axis('image')
title('Watts-Strogatz')

subplot(1, 3, 3)
nx.draw(ba)
axis('image')
title('Barabasi-Albert')

show()
```
