```python
from pylab import *
import networkx as nx
```

```python
for i,p in[(1,0.0001),(2,0.001),(3,0.01),(4,0.1)]:
    subplot(1,4,i)
    title('p='+ str(p))
    g=nx.erdos_renyi_graph(100,p)
    nx.draw(g,node_size=10)
```

```python
for i,p in[(1,0.0001),(2,0.001),(3,0.01),(4,0.1)]:
    subplot(1,4,i)
    title('p='+ str(p))
    g=nx.erdos_renyi_graph(100,p)
    nx.draw(g,node_size=10)
```

```python
p=0.00001
pdata=[]
gdata=[]
while p<0.1:
    pdata.append(p)
    g=nx.erdos_renyi_graph(100,p)
    ccs=nx.connected_components(g)
    gdata.append(max(len(cc) for cc in ccs))
    p*=1.1
```

```python
loglog(pdata,gdata)
xlabel('p')
ylabel('tamaño del cluster gigante')
show()
```
