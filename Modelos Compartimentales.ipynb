# Modelos compartimentales en epidemiología


Se parte de que es posible particionar a la población de acuerdo a su estado de salud, Por ejemplo, entre quienes están Infectados y aquellos que son susceptibles de quedar infectados con algún patógeno. En escencia el estado de salud es la variable que describe la dinámica de un sistema epidémico.



En este caso estamos suponiendo que los susceptibles  se infectan con cierta probabilidad al estar en contacto con un infectado-infeccioso y al transcurrir su enfermedad, vuelven al estado susceptible, es decir el patógeno no genera inmunidad, a este sistema se le llama modelo SIS



Ahora supongamos que, después del curso de la enfermedad el individuo que estuvo infectado desarrolla  cierta inmunidad al patógeno, en este caso la población  inmune, debe ser considerada como otro estrato o compartimento. Observe que aquí hemos incluido  Este sistema

Ambas formas de describir la dinámica infecciosa subyacen al menos tres hipótesis fuertes:

1. Hipótesis de la población bien-mezclada. Es decir que el evento de tener contacto efectivo con un infeccioso es equiprobable

2. Hipótesis de la infección instantánea. Es decir que una vez que el patógeno tiene contacto con el individuo,esteresulta infectado y pasa a ser infeccioso.

3. Hipótesis del balance demográfico. La población muere debido a la enfermedad con una tasa mas lenta que los efectos demógráficos.
